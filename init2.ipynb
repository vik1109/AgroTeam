```python
illnes_list = ['БОЛЕЗНЬ', 'МАСТИТ', 'МЕТРИТ', 'ПАРЕЗ', 'ПОСЛЕД', 'ILLMISC', 'KETOS', 'MAST', 'METR']
defect_list = ['ДЕФЕКТ', 'ХРОМОТА', 'LAME']
mastit_list = ['МАСТИТ', 'MAST']
```

```python
df_new['была_болезнь'] = df_new.событие.apply(lambda x: 1 if x in illnes_list else 0)
df_new['был_дефект'] = df_new.событие.apply(lambda x: 1 if x in defect_list else 0)
df_new['родитель_потомок'] = df_new.номер_матери.apply(lambda x: 1 if x==0 else 0) # 1 мать, 0 потомок
df_new['был_мастит'] = df_new.событие.apply(lambda x: 1 if x in mastit_list else 0)
```

```python
# мастит у матери
mother_mastit = pd.DataFrame(df_new[(df_new.был_мастит==1) & (df_new.родитель_потомок==1)]
                             .groupby(['код_отца'])['номер_животного'].count())
mother_mastit.rename(columns={'номер_животного': 'кол_во_матерей_с_маститом'}, inplace=True)
mother_mastit
```

```python
# мастит у потомка
child_mastit = pd.DataFrame(df_new[(df_new.был_мастит==1) & (df_new.родитель_потомок==0)]
                             .groupby(['код_отца'])['номер_животного'].count())
child_mastit.rename(columns={'номер_животного': 'кол_во_детей_от_матерей_с_маститом'}, inplace=True)
child_mastit
```

# Сборка исследовательского датасета по отцам, матерям, потомкам

```python
df_father_research = df_father.copy()

df_father_research['больных_потомков(все болезни)'] = df_new[df_new.была_болезнь==1]\
           .groupby(['код_отца'])['номер_животного'].count()
df_father_research['больных_потомков_с_маститом'] = df_new[df_new.был_мастит==1]\
           .groupby(['код_отца'])['номер_животного'].count()
df_father_research['кол_во_матерей_с_маститом'] = df_new[(df_new.был_мастит==1) & (df_new.родитель_потомок==1)]\
                             .groupby(['код_отца'])['номер_животного'].count()
df_father_research['кол_во_потомков_от_матерей_с_маститом'] = df_new[(df_new.был_мастит==1) & (df_new.родитель_потомок==0)]\
                             .groupby(['код_отца'])['номер_животного'].count()
df_father_research.fillna(0, inplace=True)

df_father_research = df_father_research.astype(int)

df_father_research
```

Обогащение исследовательского датасета фичами

```python
df_father_research['%_больного_потомства'] = round(df_father_research['больных_потомков(все болезни)']\
                                                / df_father_research.всего_потомков * 100, 2)
df_father_research['%_потомства_с_маститом(от общего числа потомков)'] = round(df_father_research.больных_потомков_с_маститом\
                                                / df_father_research.всего_потомков * 100, 2)
df_father_research['%_потомства_с_маститом(от общего числа заболевавших)'] = round(df_father_research.больных_потомков_с_маститом\
                                                / df_father_research['больных_потомков(все болезни)'] * 100, 2)

df_father_research['%_матерей_с_маститом(от общего числа потомков)'] = round(df_father_research.кол_во_матерей_с_маститом\
                                                / df_father_research.всего_потомков * 100, 2)
df_father_research['%_матерей_с_маститом(от общего числа заболевавших)'] = round(df_father_research.кол_во_матерей_с_маститом\
                                                / df_father_research['больных_потомков(все болезни)'] * 100, 2)
df_father_research['%_матерей_с_маститом(от общего числа заболевших маститом)'] = round(df_father_research.кол_во_матерей_с_маститом\
                                                / df_father_research['больных_потомков_с_маститом'] * 100, 2).fillna(0)

df_father_research['%_потомков_с_маститом(от общего числа потомков)'] = round(df_father_research.кол_во_потомков_от_матерей_с_маститом\
                                                / df_father_research.всего_потомков * 100, 2)
df_father_research['%_потомков_с_маститом(от общего числа заболевавших)'] = round(df_father_research.кол_во_потомков_от_матерей_с_маститом\
                                                / df_father_research['больных_потомков(все болезни)'] * 100, 2)
df_father_research['%_потомков_с_маститом(от общего числа заболевших маститом)'] = round(df_father_research.кол_во_потомков_от_матерей_с_маститом\
                                                / df_father_research['больных_потомков_с_маститом'] * 100, 2).fillna(0)
df_father_research['%_потомков_с_маститом(от матерей болевших маститом)'] = round(df_father_research.кол_во_потомков_от_матерей_с_маститом\
                                                / df_father_research.кол_во_матерей_с_маститом * 100, 2).fillna(0)


df_father_research
```

"-" - означает, что корова была куплена
<br><br>
По купленным коровам наблюдается взаимосвязь болезни матери и потомка. Всего подобных коров переболело маститом 245, из них 125 коров дали потомство и 120 потомков болело маститом
<br><br>
В остальных случаях, не всегда явно проявляется взаимосвязь факт болезни (мастит) матери с аналогичной болезнью потомка
<br><br>
Выявлено, что от некоторых отцов рождается больше коров склонных к заболеваниям. График лидеров приведен ниже

```python
df_father_research.plot.scatter(x='%_больного_потомства', y='%_потомства_с_маститом(от общего числа потомков)',
                               figsize=(10,8))
plt.title('Соотношение доли заболевшего потомства (различными болезнями) с долей заболевших маститом коров')
plt.xlabel('% больного потомства')
plt.ylabel('% потомков с маститом')
plt.show()
```
